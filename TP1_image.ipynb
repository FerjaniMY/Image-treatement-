### I.2. Types des données (images)

On trouve plusieurs conventions pour répresenter les valeurs des pixels d'une image:

```
  0 - 255   avec  0 est noir, 255 est blanc
  0 - 1     avec  0 est noir, 1 est blanc
```

Le type de l'image est déterminé par ``.dtype`` .

On prend l'exemple du code ci-dessous permettant de générer deux images en niveaux de gris ayant deux différents types:

````python
# exemple 1
coins.dtype#retourne le type de données
````

````python
# exemple 2

import numpy as np
from matplotlib import pyplot as plt

image0 = np.linspace(0, 1, 900).reshape((30, 30))#'reshape pour reajuster la taille du matrice'
# linspace :genere un vecteur de 900 valeurs entre 0 et 1
image1 = np.linspace(0, 255, 900).reshape((30, 30)).astype(np.uint8)#astype : pour definir le type de "data"

print("image0:", image0.dtype, image0.min(), image0.max())
print("image1:", image1.dtype, image1.min(), image1.max())

fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(5, 5))
ax0.imshow(image0, cmap='gray')
ax1.imshow(image1, cmap='gray');
image1
````

#### Exercice 3: Convertir une image RVB en HSV

La transformation Teinte saturation luminance (en Anglais, HSV) est l'une des transformations les plus connues dans les espaces des couleurs. La commande ``color.rgb2hsv`` permet de passer d'un espace RVB par exemple en espace HSV.

* Lire l'image "lena.jpg" et l'afficher en couleurs RVB.
* Convertir l'image Lena de l'espace RVB vers l'espace HSV.
* Séparer les trois canaux corresponsant à la Teinte (H), saturation (S) et luminance (V) et les afficher dans la meme figure. 
* Sauvegarder chaque canal (H, S, V) sous format ".jpg".


````python
from skimage import io,color

# charger l'image Lena dans une variable X
X = io.imread('images/lena.jpg')

# Voir les caractéristiques de l'image Lena

print(type(X))
print(X.dtype)
print(X.shape)
print(X.min(), X.max())

# Afficher la matrice de l'image X 
plt.imshow(X);

# Sauvegarder l'image X dans une nouvelle image avec un autre nom
plt.imsave('images/image_lena.jpg', X) # save image to disk

````

````python
X = io.imread('images/lena.jpg')
y=color.rgb2hsv(X)#Convertir l'image Lena de l'espace RVB vers l'espace HSV.
plt.imshow(y)
````

````python
H=y[:, :, 0]
plt.imsave('images/H_lena.jpg',H )

S=y[:,:,1]
plt.imsave('images/S_lena.jpg',S )

V=y[:, :, 2]
plt.imsave('images/V_lena.jpg',V )

plt, axes = plt.subplots(1, 3)
axes[0].imshow(H)
axes[1].imshow(S)
axes[2].imshow(V)
````

## II. Histogramme 

L'histogramme d'une image est la distribution
La fonction ``hist(image.ravel(), lw=0, bins=256)`` permet de calculer l'histogramme d'une image. L'histogramme de l'image 'image' en utilisant un nombre 'bins' niveaux de gris. Elle compte le nombre de pixels par niveaux de gris et l'afficher . 
Par exemple:

```python

#charger l'image 
image = data.camera()

# Afficher l'image
fig, (ax1, ax2) = plt.subplots(1, 2,
                              figsize=(10, 3))
ax1.imshow(image, cmap='gray')
ax1.set_axis_off()

# Afficher l'histogramme
ax2.hist(image.ravel(), lw=0, bins=256)
ax2.set_xlim(0, image.max())
ax2.set_yticks([])

plt.show()

```

**Question 6:** Tester l'exemple ci-dessus.

````python
image = data.camera()

# Afficher l'image
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(10, 3))
ax1.imshow(image, cmap='gray')
ax1.set_axis_off()

# Afficher l'histogramme
ax2.hist(image.ravel(), lw=0, bins=256)
ax2.set_xlim(0, image.max())
ax2.set_yticks([])

plt.show()
````
