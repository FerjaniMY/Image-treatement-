```python
camerae=skimage.morphology.erosion(camera,carre)
camerad=skimage.morphology.dilation(camera,carre)
```

```python
fig=plt.figure
plt.subplot(1,3,1)
plt.imshow(camera,cmap='gray')
plt.title('image originale')
plt.subplot(1,3,2)
plt.imshow(camerae,cmap='gray')
plt.title('image avec rision')
plt.subplot(1,3,3)
plt.imshow(camerad,cmap='gray')
plt.title('image avec dilatation')
```
